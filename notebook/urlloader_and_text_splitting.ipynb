## URL LOADER CLASS FROM LANGCHIAN ##


```python
from langchain_community.document_loaders import UnstructuredURLLoader

loader = UnstructuredURLLoader(urls = ["https://blog.python.org/"])
data = loader.load()

data[0].page_content
```

## TEXT SPLITTING INTO CHUNKS ##

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


splitted_text = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " "],
    chunk_size=200,  
)


docs = splitted_text.split_text(data[0].page_content)

docs  # list which contain chunks in the form of list.
```

```python
# converting this list of strings to list of documents because embeddings work on document format.

from langchain_core.documents import Document

document_format = []
for doc in docs :
    document_format.append(Document(page_content = doc))

document_format
```
