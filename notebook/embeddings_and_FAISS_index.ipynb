## EMBEDDING ##
Embeddings are a way to convert human-readable data like text into a format that machines can understand, which is numbers. Since computers cannot directly understand language, embeddings transform words, sentences, or documents into vectors (lists of numbers). These vectors are designed in such a way that they capture the meaning and context of the text.

For example, words or sentences with similar meanings will have vectors that are close to each other in a mathematical space, while unrelated words will be far apart. This property makes embeddings extremely useful for tasks like searching, recommendation systems, and chatbots.

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

text = "This is a test document."

query_result = embeddings.embed_query(text)

query_result
```

## FAISS INDEX (A VECTOR STORE) ## 
When we convert text into embeddings, we get vectors (lists of numbers). If we have thousands or millions of such vectors, searching through all of them to find similar ones becomes very slow. This is where a FAISS index is used.

A FAISS index organizes and stores these vectors in a way that allows very fast similarity search. Instead of comparing a query vector with every single vector one by one, FAISS uses optimized algorithms to quickly find the nearest (most similar) vectors. This makes it highly efficient and suitable for real-world applications like chatbots, recommendation systems, and search engines.

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS


# gives an index to the query "hello world" based on the dimension of the embedding.
# IndexFlatL2 is a method of faiss class which helps in finding the closest vectors to the query vector.
index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))


# creating a store for the vector database
vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)

```

```python
from uuid import uuid4
# uuid is a module to generate random unique id.

from langchain_core.documents import Document

document_1 = Document(
    page_content="I had chocolate chip pancakes and scrambled eggs for breakfast this morning.",
    metadata={"source": "tweet"},
)

document_2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
)

document_3 = Document(
    page_content="Building an exciting new project with LangChain - come check it out!",
    metadata={"source": "tweet"},
)

document_4 = Document(
    page_content="Robbers broke into the city bank and stole $1 million in cash.",
    metadata={"source": "news"},
)

document_5 = Document(
    page_content="Wow! That was an amazing movie. I can't wait to see it again.",
    metadata={"source": "tweet"},
)

document_6 = Document(
    page_content="Is the new iPhone worth the price? Read this review to find out.",
    metadata={"source": "website"},
)

document_7 = Document(
    page_content="The top 10 soccer players in the world right now.",
    metadata={"source": "website"},
)

document_8 = Document(
    page_content="LangGraph is the best framework for building stateful, agentic applications!",
    metadata={"source": "tweet"},
)

document_9 = Document(
    page_content="The stock market is down 500 points today due to fears of a recession.",
    metadata={"source": "news"},
)

document_10 = Document(
    page_content="I have a bad feeling I am going to get deleted :(",
    metadata={"source": "tweet"},
)

documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]

uuids = []

# generating random unique id for each document
for _ in range(len(documents)) :
    uuids.append(str(uuid4()))    # str() method converts unique id into string.


# adding each of these document with their uuid to the vector store.
vector_store.add_documents(documents=documents, ids=uuids)
```

```python
# performing a search

result = vector_store.similarity_search("what is the latest news?")

for doc in result:
    print(doc.page_content)
```

```python
# saving and loading faiss index 

# saving the data locally into disc so that we can retrive it whenever we want. 
vector_store.save_local("rough_faiss_index")

new_vector_store = FAISS.load_local(
    "rough_faiss_index", embeddings, allow_dangerous_deserialization=True
)
```

```python
docs1 = new_vector_store.similarity_search("what is the latest news?")

docs1[0].page_content
```

## Vector Store Retriver ## 
A Vector Store Retriever in LangChain is the most common type of retriever that lets you
search and fetch documents from a vector store based on semantic similarity using vector
embeddings.

NOTE :- Retriver performs advanced searches hence we use them instead of 'similarity_search()' method.

```python
# convert vector store into retriver

retriver = vector_store.as_retriever(search_kwargs={'k':2}) # means get top 2 results

results = retriver.invoke("apple iphone latest updates")

results

for i, res in enumerate(results):
    print(f"Result -- {i+1}--\n")
    print(res.page_content)
    print("\n")
```
